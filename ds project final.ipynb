# Sorting and Viewing the DataFrames

```python
# sorting the data frames by province/state and country/region.

df_confirmed = df_confirmed.sort_values(by=['Province/State', 'Country/Region'])
df_confirmed = df_confirmed.reset_index(drop=True)

df_death = df_death.sort_values(by=['Province/State', 'Country/Region'])
df_death = df_death.reset_index(drop=True)

df_recovered  = df_recovered.sort_values(by=['Province/State', 'Country/Region'])
df_recovered = df_recovered.reset_index(drop=True)

# let's have a look, how a dataframe looks like.
df_confirmed
```

# Extract the Names of Columns Containing Dates and Times

```python
dates_confirmed = df_confirmed.columns[4:]
dates_death = df_death.columns[4:]

print(dates_confirmed)
print(dates_death)

# make sure the dates are all the same for the 2 dfs
if dates_confirmed.equals(dates_death):
    print("Dates are the same")
```

# Unpivot a DataFrame From Wide Format to Long Format

```python
# unpivot the dataframes so that the dates are no longer represented as columns.
# the dates should be stored as values under a new column, say Date. 
# The number of cases (Confirmed, Deaths) should be saved as a corresponding column, say Confirmed, and Deaths. 
# For this, I can use the melt() function:

df_confirmed_melted = df_confirmed.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                        value_vars=dates_confirmed,
                                       var_name='Date',
                                       value_name='Confirmed')

df_death_melted = df_death.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                        value_vars=dates_death,
                                       var_name='Date',
                                       value_name='Death')

```

# View the Unpivoted DataFrames

```python
df_confirmed_melted
```

```python
df_death_melted
```

```python
# to make sure that the two dataframes have the same shape.

print(df_confirmed_melted.shape)
print(df_death_melted.shape)
```

# Combine All the Unpivoted Dataframes into One Single Dataframe

```python
df_combined = pd.concat([df_confirmed_melted,
                        df_death_melted["Death"]],
                       axis=1,
                       sort=False)
df_combined
```

# calculating percentage of deaths with respect to number of confirmed cases

```python
df_combined["Percentage"] = df_combined["Death"]/df_combined["Confirmed"]

df_combined
```

```python
# Because there are cells that are NaNs (caused by division-by-zero), it is important to replace them with 0s:

df_combined.fillna(0, inplace=True)
df_combined
```

```python
# it is useful to check the data types for the various columns in the dataframe

df_combined.dtypes
```

```python
# we can see that the date column is of type object.
```

# Change the Date column to datetime Format

```python
import datetime

def format_time(datetime_str):
    d=datetime.datetime.strptime(datetime_str, '%m/%d/%y')
    return d.date()

df_combined["Date"] = df_combined["Date"].apply(format_time)
df_combined
```

# Display the Daily Number of Confirmed and Deaths Cases

```python
# Now that the dataframes have been combined into one and significantly cleaned, we can start performing some 
# analytics so that we can derive some useful information from the dataframe.
# Let’s start by displaying the daily number of confirmed and death cases:
```

```python
grouped_by_date = df_combined.groupby("Date")
for date, group in grouped_by_date:
    print("==========")
    print(date)
    print("==========")
    print(group[["Province/State","Country/Region", "Confirmed", 
        "Death"]])
    
# You will now see the number of confirmed and death cases for each day:
```

# Display the Total Daily Number of Confirmed and Deaths For Each Country

```python
# The previous result showed the number for each province/state in each country. 
# For the total numbers for each country, perform an aggregation:
```

```python
df_daily = df_combined.groupby(["Date", "Country/Region"]).aggregate({'Confirmed': 'sum', 'Death':'sum'})
df_daily
```

# Get the Data For a Specific Day

```python
# use the loc[] indexer to get the numbers for a particular day.

df_day = df_daily.loc[format_time('3/22/20')]
df_day
```

# Get the Data For a Specific Day For a Specific Country

```python
# we can retrieve the numbers for a paarticular country:

df_day_country = df_daily.loc[(format_time('5/22/20'), 'India')]
df_day_country
```

# Display the Data For The Most Recent Day

```python
# To get the numbers for the most recent day, first get the last date 
# in the dataframe and then use it to extract the data for that date:

# get the most recent date

most_recent_date = df_combined['Date'].max()
print(most_recent_date)
```

```python
# get all the data on the most recent date

df_most_recent = df_daily.loc[most_recent_date, :]
df_most_recent
```

# Top 10 Countries with Confirmed Cases

```python
# To get the top 10 countries with confirmed cases, sort the df_most_recent dataframe by the 
# Confirmed column and then display the first 10 rows:


df_most_confirmed_recent_sorted = df_most_recent.sort_values(by="Confirmed", ascending=False)
df_most_confirmed_recent_sorted.head(10)
```

# Top 10 Countries with Most Deaths Reported

```python
# To get the top 10 countries with deaths reported, first sort the df_most_recent dataframe using the Deaths column. 
# Then, filter out those countries that has no death reported. 
# Finally, show the first 10 rows:

df_most_death_recent_sorted = df_most_recent.sort_values(by="Death", ascending=False)

df_most_death_recent_sorted = df_most_death_recent_sorted [df_most_death_recent_sorted.Death>0]

df_most_death_recent_sorted.head(10)
```

# Display the Most Recent Data for Mainland China

```python
# Back in the early days of Covid-19, quite a number of provinces in China were infected. 
# And so it would be interesting see the numbers for the various provinces in China:

# get all the rows for Mainland China
df_combined_china = df_combined[df_combined['Country/Region'] == "China"]
    
# using the china dataframe, group by date and province
df_combined_china_date = df_combined_china.groupby(["Date","Province/State"]).aggregate(
        {'Confirmed': 'sum', 'Death': 'sum'})

# find the data for the most recent date
df_combined_china_most_recent = df_combined_china_date.loc[most_recent_date,:]

df_combined_china_most_recent
```

# Display the Top 10 Provinces in China With Confirmed Cases

```python
# To get the top 10 provinces in China with the most confirmed cases, 
# sort the df_combined_china_most_recent dataframe by the Confirmed column and returns its top 10 rows:

df_combined_china_most_recent_sorted = df_combined_china_most_recent.sort_values(by="Confirmed", ascending=False).head(10)
df_combined_china_most_recent_sorted

# As you can observe, Hubei is the most affected province in China.
```
